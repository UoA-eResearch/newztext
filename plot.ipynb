```python
import pandas as pd # Loading/filtering data
import matplotlib.pyplot as plt # plotting
import seaborn as sns # plot style
sns.set()
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
df = pd.read_html("they_are_us.html")[0]
df
```

```python
df.Source.value_counts()
```

```python
df.Date = pd.to_datetime(df.Date)
df.index = df.Date
for interval in ["day", "week", "month"]:
    df.Date[df.Source.isin(["New Zealand Herald", "New Zealand Herald Sat", "Herald on Sunday"])].resample(interval[0]).count().plot(title="Articles mentioning they are us")
    df.Date[df.Source.isin(["THE PRESS", "STUFF — THE PRESS"])].resample(interval[0]).count().plot()
    plt.legend(["NZ Herald", "The Press"])
    plt.ylabel('# of articles per ' + interval)
    plt.show()
```
